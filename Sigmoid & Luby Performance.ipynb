```python
from mighty.utils.logger import Logger, load_logs, log2dataframe
import pandas as pd
#we may also want to migrate and extend that at some point or just build something completely different
from dacbench.plotting import plot_performance, plot_performance_per_instance
```

```python
def load_results(path, seeds):
    train_results = []
    eval_results = []
    for seed in seeds:
        logs = load_logs(f"{path}_s{seed}/train_performance.jsonl")
        train_results.append(log2dataframe(logs, wide=True))
        logs = load_logs(f"{path}_s{seed}/eval_performance.jsonl")
        eval_results.append(log2dataframe(logs, wide=True))
    train_results = pd.concat(train_results)
    eval_results = pd.concat(eval_results)
    return train_results, eval_results
```

```python
path = "sigmoid/sigmoid_example"
seeds = [0, 1, 2, 3, 4]
train_sigmoid, evaluation_sigmoid = load_results(path, seeds)
```

```python
plot = plot_performance(train_sigmoid, title="Train Reward on Sigmoid Benchmark (with 95% CI)", aspect=2)
```

```python
plot = plot_performance(train_sigmoid, title="Train Reward per seed on Sigmoid Benchmark (with 95% CI)", hue="seed", aspect=2)
```

```python
plot_performance_per_instance(evaluation_sigmoid, title="Evaluation Performance per training instance (Sigmoid)", aspect=2)
```
